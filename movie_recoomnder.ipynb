```python
def clean_text(text):
  tokens=word_tokenize(str(text).lower()) # for lower case
  tokens=[t for t in tokens if t.isalpha()] # for only words
  tokens=[t for t in tokens if t not in stopwords.words("english")]
  return ' '.join(tokens)


```

```python
df["cleaned_descrip"]=df["description"].apply(clean_text)
```
